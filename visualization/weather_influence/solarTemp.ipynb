```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('../../allMergedCleaned.csv')

# Convert 'Tidspunkt' to datetime format
df['Tidspunkt'] = pd.to_datetime(df['Tidspunkt'])

# Filter data for building 10724
df_10724 = df[df['property_id'] == 10724]

# Plotting solar consumption against temperature
plt.figure(figsize=(10, 6))
plt.scatter(df_10724['temperature'], df_10724['solar_consumption'], color='#66b3ff', alpha=0.7)
plt.title('Solar Consumption vs Temperature for Building 10724')
plt.xlabel('Temperature (°C)')
plt.ylabel('Solar Consumption (kW)')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Plotting solar consumption against temperature with cloud fraction as color
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    df_10724['temperature'],
    df_10724['solar_consumption'],
    c=df_10724['cloud_fraction'],
    cmap='coolwarm',  # Color map representing cloud cover intensity
    alpha=0.7
)
plt.title('Solar Consumption vs Temperature for Building 10724')
plt.xlabel('Temperature (°C)')
plt.ylabel('Solar Consumption (kW)')

# Adding color bar for cloud fraction
cbar = plt.colorbar(scatter)
cbar.set_label('Cloud Fraction')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Select only the relevant columns
correlation_data = df_10724[['solar_consumption', 'temperature', 'cloud_fraction']]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Map for Solar Consumption, Temperature, and Cloud Fraction')
plt.show()
```
