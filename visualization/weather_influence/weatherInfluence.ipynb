Load data and process it

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('../../allMergedCleaned.csv')  # Replace with your actual path

# Convert 'Tidspunkt' to datetime if not already done
data['Tidspunkt'] = pd.to_datetime(data['Tidspunkt'])

# Set 'Tidspunkt' as the index for time-series analysis, if needed
data.set_index('Tidspunkt', inplace=True)
```

Correlation Matrix Heatmap

```python
# Calculate correlation matrix
corr = data[['Verdi', 'temperature', 'wind_speed', 'cloud_fraction', 'precipitation']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix of Energy Consumption and Weather Factors")
plt.show()
```

Scatter Plots for Key Weather Variables

```python
# Define subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Scatter plot for temperature vs energy consumption
sns.scatterplot(ax=axes[0, 0], x='temperature', y='Verdi', data=data)
axes[0, 0].set_title('Temperature vs Energy Consumption')
axes[0, 0].set_xlabel('Temperature (°C)')
axes[0, 0].set_ylabel('Energy Consumption (KW)')

# Scatter plot for wind speed vs energy consumption
sns.scatterplot(ax=axes[0, 1], x='wind_speed', y='Verdi', data=data)
axes[0, 1].set_title('Wind Speed vs Energy Consumption')
axes[0, 1].set_xlabel('Wind Speed (m/s)')
axes[0, 1].set_ylabel('Energy Consumption (KW)')

# Scatter plot for cloud fraction vs energy consumption
sns.scatterplot(ax=axes[1, 0], x='cloud_fraction', y='Verdi', data=data)
axes[1, 0].set_title('Cloud Fraction vs Energy Consumption')
axes[1, 0].set_xlabel('Cloud Fraction')
axes[1, 0].set_ylabel('Energy Consumption (KW)')

# Scatter plot for precipitation vs energy consumption
sns.scatterplot(ax=axes[1, 1], x='precipitation', y='Verdi', data=data)
axes[1, 1].set_title('Precipitation vs Energy Consumption')
axes[1, 1].set_xlabel('Precipitation (mm)')
axes[1, 1].set_ylabel('Energy Consumption (KW)')

plt.tight_layout()
plt.show()
```
